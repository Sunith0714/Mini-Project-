# Univariate Analysis for Boston Housing

>Univariate analysis involves examining individual variables or features in isolation to gain insights into their distributions, characteristics, and potential impact on the target variable. In the context of the Boston Housing dataset, univariate analysis can help you understand each feature's properties. Here's how to perform univariate analysis for the dataset using Python and some common visualization techniques:

# 1. Univariate Analysis: Histograms

# Create a histogram for each feature

```python
plt.figure(figsize=(15, 10))
for i, feature in enumerate(boston_data.columns):
    plt.subplot(4, 4, i + 1)
    sns.histplot(boston_data[feature], bins=30, kde=True)
    plt.title(f'Histogram of {feature}')
plt.tight_layout()
plt.show()
```

# Bivariate Analysis for Boston housing 

# 1. Bivariate Analysis: Scatter Plots

Create scatter plots to visualize the relationship between selected features and the target variable (MEDV)


```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
sns.scatterplot(x='rm', y='medv', data=boston_data)
plt.title("Scatter Plot of RM (Average Number of Rooms) vs. medv")
plt.subplot(1, 2, 2)
sns.scatterplot(x='lstat', y='medv', data=boston_data)
plt.title("Scatter Plot of LSTAT (Percentage of Lower Status Population) vs. medv")
plt.show()
```

Create a pairwise scatter matrix to visualize the relationships between multiple features

```python
sns.pairplot(boston_data, vars=['rm', 'lstat', 'crim', 'tax', 'medv'])
plt.show()
```

# 3. Bivariate Analysis: Correlation Heatmap

Create a correlation heatmap to visualize the pairwise correlations between features


```python
correlation_matrix = boston_data.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

# Multivariate Analysis for Boston Housing dataset

# Multivariate Analysis: 3D Scatter Plots
Create 3D scatter plots to explore interactions between three features and the target variable

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(boston_data['rm'], boston_data['lstat'], boston_data['medv'], c='b', marker='o')
ax.set_xlabel('rm')
ax.set_ylabel('lstat')
ax.set_zlabel('medv')
plt.title("3D Scatter Plot: RM, LSTAT, and MEDV")
plt.show()
```

# Multivariate Analysis: Parallel Coordinates Plot

Create a parallel coordinates plot to visualize the relationships between multiple features and the target variable


```python
from pandas.plotting import parallel_coordinates

plt.figure(figsize=(18, 14))
parallel_coordinates(boston_data[['rm', 'lstat', 'crim', 'tax', 'medv']], 'medv')
plt.title("Parallel Coordinates Plot")
plt.show()
```

# Visualize above with Bar, Box, FacetGrid, Heatmap, Crosstab, Line plot

# Bar Plot for 'CHAS' (Charles River dummy variable)

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='chas', data=boston_data)
plt.title("Bar Plot for CHAS")
plt.xlabel("CHAS")
plt.ylabel("Count")
plt.show()
```

```python
# Box Plot for 'RM'
plt.figure(figsize=(6, 4))
sns.boxplot(x='rm', data=boston_data)
plt.title("Box Plot for RM")
plt.xlabel("RM")
plt.show()

# FacetGrid
g = sns.FacetGrid(boston_data, col='chas')
g.map(plt.hist, 'rm', bins=20)
g.set_axis_labels("rm", "Frequency")
g.fig.suptitle("FacetGrid for RM by CHAS", y=1.02)
plt.show()

# Line Plot (Trends over data points)
plt.figure(figsize=(8, 6))
sns.lineplot(x=boston_data.index, y='rm', data=boston_data)
plt.title("Line Plot for RM")
plt.xlabel("Data Point")
plt.ylabel("RM")
plt.show()
```
